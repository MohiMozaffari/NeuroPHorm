```python
import os
from pathlib import Path
import numpy as np
os.chdir(r"C:\Users\Mohad\Documents\Programming\NeuroPHorm")
import neurophorm as nf
```

```python
OUTDIR_NODE = Path("examples/node_removal") 
OUTDIR_PERSIST = Path("examples/persistences") 
```

```python
np.random.seed(42)
atlas = np.array(["I"* i for i in np.random.randint(1, 4, size=25)])
```

```python
mean_df, error_df  = nf.load_removal_data(OUTDIR_NODE, atlas)
data = nf.load_tda_results(
        output_directory=OUTDIR_PERSIST,
        load_all=False,
        load_betti=True,
        load_entropy=True,
        load_amplitude=True,
        include_all_betti_x=True,
        interpolate_betti_to_shared_x=True,
    )
betti = nf.compute_betti_stat_features(data)
```

```python
res = nf.collect_pvalue_counts(
    data=data,
    alpha=0.05, multitest="bonferroni",
    betti_stats=betti,  
    node_mean=mean_df, 
    node_se=error_df,
    atlas=atlas)
```
